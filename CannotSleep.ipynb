```python
import numpy as np
import random
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sleep 
#p not gettling sleep (assumes once a week awake sleep)
awake = (1/28)

#awake hours are between 1A-5A and awake is for 30 min at a time
awake = awake / 10
```

```python
def both_awake():
    awake_andrew = np.random.binomial(1, awake)
    awake_carolyn = np.random.binomial(1, awake)
    if awake_andrew == 1 and awake_carolyn == 1:
        WeAreBothAwake = True
    else:
        WeAreBothAwake = False
    return WeAreBothAwake
```

```python
probability_awake = []
total_awakes = []
total_simulations = 10000

for i in range(total_simulations):
    days_per_year = 365
    years_together = 9
    periods_night = 10
    totalsleeptimes = days_per_year*years_together*periods_night
    n_awake=0
    total_sleep_periods=[0]

    while total_sleep_periods[-1] < total_simulations:
        cantSleep = both_awake()
        if cantSleep == True:
            n_awake += 1
            total_sleep_periods.append(total_sleep_periods[-1] + 1)
        else:
            total_sleep_periods.append(total_sleep_periods[-1] + 1)
    
    probability_awake.append(np.sum(n_awake) / len(total_sleep_periods))
    total_awakes.append(np.sum(n_awake))
```

```python
print("Probability for any one night Carolyn and Andrew wake up at the same time in the middle of the night is low with average p =", np.mean(probability_awake))
awakePlot = sns.histplot(probability_awake)
print(awakePlot)
```

```python
print("Average awake events over the course of relationship ", np.mean(total_awakes), "times.")
hist = sns.histplot(total_awakes)
hist.set(xlabel='Number of Times Awake at Same Time', ylabel='n')
print(hist)
```
